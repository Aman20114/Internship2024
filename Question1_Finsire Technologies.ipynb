```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()

# Define target cities
cities = ['Hyderabad', 'Jaipur', 'Visakhapatnam', 'Chandigarh', 'Rajkot', 'Surat', 'Madurai', 'Thiruvananthapuram', 'Indore', 'Lucknow']

# Empty list to store data
real_estate_data = []

# Loop through each city
for city in cities:
    url = f"https://www.99acres.com/property-in-{city.lower()}-ffid"
    driver.get(url)
    
    try:
        # Wait for the locations to be present
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[@class="T_091c165f _sq1l2s _vv1q9c _ks15vq T_3d3547ab _7s5wglyw _5vy24jg8 _blas1v10 new-title"]/span')))
        
        # Scrape data
        locations = driver.find_elements(By.XPATH, '//div[@class="T_091c165f _sq1l2s _vv1q9c _ks15vq T_3d3547ab _7s5wglyw _5vy24jg8 _blas1v10 new-title"]/span')
        prices_per_sqft = driver.find_elements(By.XPATH, '//div[@class="T_091c165f _sq1l2s _vv1q9c _ks15vq T_efe231cd _vy1ipv _7ltvct _g3dlk8 _c81fwx _cs1nn1 value"]')
        rental_yields = driver.find_elements(By.XPATH, '/html/body/div[1]/div[2]/div[2]/div[2]/div[1]/div[3]/div[1]/article/div[1]/div[2]/div[1]/div[3]/section/div[1]/div[2]/div[2]/div[2]')
        
        for i in range(len(locations)):
            data = {
                'Location': locations[i].text,
                'City': city,
                'State': 'Unknown',  # Update with actual state if available
                'Price Per Sq Ft': prices_per_sqft[i].text if i < len(prices_per_sqft) else 'N/A',
                'Rental Yield': rental_yields[i].text if i < len(rental_yields) else 'N/A'
            }
            real_estate_data.append(data)
    except Exception as e:
        print(f"Error scraping data for {city}: {e}")

# Close the WebDriver
driver.quit()

# Convert to DataFrame and save to CSV
df = pd.DataFrame(real_estate_data)
df.to_csv('real_estate_data.csv', index=False)

print("Data scraping complete and saved to real_estate_data.csv")
```
