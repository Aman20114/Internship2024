```python
driver = webdriver.Chrome()

driver.get("https://github.com/explore")
```

```python
trending_page = driver.find_element(By.XPATH, '/html/body/div[1]/div[4]/main/div[1]/nav/div/a[3]')

trending_page.click()
```

```python
repos = []
url = driver.find_elements(By.XPATH, '//a[@class="Link"]')
for i in url:
   repos.append(i.get_attribute('href'))

time.sleep(2)
```

```python
Repository_titles = []
Repository_descriptions = []
Contributors_counts = []
Language_uses = []
```

```python
for url in repos:
    driver.get(url)# This will open every product link individually
    time.sleep(3)
    
    try:
        title = driver.find_element(By.XPATH, '/html/body/div[1]/div[4]/div/main/div/div[1]/div[1]/div')
        Repository_titles.append(title.text)
    except:
        Repository_titles.append('_')
    try:
        description = driver.find_element(By.XPATH, '//p[@class="col-9 color-fg-muted my-1 pr-4"]')
        Repository_descriptions.append(description.text) 
    except:
        Repository_descriptions.append('_')
    try:
        count = driver.find_element(By.XPATH, '/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[5]/div/h2/a')
        Contributors_counts.append(count.text)
    except:
        Contributors_counts.append("_")
    try: 
        lang = driver.find_Element(By.XPATH, '/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[6]/div/ul/li[2]/a/span[1]')
        Language_uses.append(lang.text)
    except:
        Language_uses.append("_")

df = pd.DataFrame({
    "Repository title":Repository_titles, 
    "Repository description":Repository_descriptions, 
    "Contributors count":Contributors_counts, 
    "Language used":Language_uses
})
df
```

```python
driver = webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
```

```python
all_datasets_link = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[2]/ul/li[1]/a')
all_datasets_link.click()
```

```python
rows = driver.find_elements(By.XPATH, '//div[@class="rounded-box grid grid-cols-8 gap-2 p-2 hover:bg-base-200"]')

for row in rows[1:]:
    dataset_name = row.find_element(By.XPATH, '//h2[@class="truncate text-primary"]/a').text
    
    data_type = row.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div/div[2]/p/text()').text
    
    task = row.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div/div[2]/div/div[1]/span').text
    
    attribute_type = row.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div/div[2]/div/div[4]/span').text
    
    num_instances = row.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div/div[2]/div/div[4]/span').text
    
    num_attributes = row.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div/div[2]/div/div[4]/span').text
    
    year = row.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div/div[2]/div/div[4]/span').text

    print("Dataset name:", dataset_name)
    print("Data type:", data_type)
    print("Task:", task)
    print("Attribute type:", attribute_type)
    print("No of instances:", num_instances)
    print("No of attributes:", num_attributes)
    print("Year:", year)
    print()
```
